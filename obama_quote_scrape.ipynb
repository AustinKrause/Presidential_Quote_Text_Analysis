<h1>INSERT QUOTES INTO DICTIONARY</h1>

```python
full_list_quotes = []
```

```python
def insert_quotes(name, party, lst):
    for i in lst:
        all_quotes = {}
        all_quotes['said_by'] = name
        all_quotes['party'] = party
        all_quotes['quote'] = i
        full_list_quotes.append(all_quotes)
```
